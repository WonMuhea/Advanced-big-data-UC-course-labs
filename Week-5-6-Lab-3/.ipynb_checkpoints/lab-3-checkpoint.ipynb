### Final Discussion

#### Which algorithm produced better-defined clusters?

- **K-Means** **Silhouette Score** and **Adjusted Rand Index (ARI)** were both slightly higher compared to **K-Medoids** values, indicating that it formed more well-defined and separable clusters in this dataset.
- **K-Means** often benefits from smooth, convex clusters due to its reliance on Euclidean distances to centroids, which worked well here.

#### What differences are observed in cluster shapes or positioning?

- **K-Means clusters** tend to be more circular and evenly spaced because centroids are calculated using the mean of data points.
- **K-Medoids clusters**, on the other hand, may appear slightly **irregular or asymmetric**, particularly where a medoid lies on the border of a natural cluster. This happens because **K-Medoids** selects actual data points as cluster centers. That makes it more robust but sometimes less geometrically balanced.
- **K-Means** is convenient for a dataset with convex clusters, and classification speed in crucial
- **K-Medoids** fucntion better when there are outlines in the dataset, and when a distance metrics other than Euclidean needs to be used.
