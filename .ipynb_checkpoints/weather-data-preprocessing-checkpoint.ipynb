```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

<span style="font-size:20px; font-weight:bold">1. Data collection</span><br>
The data used for this assignment is a daily weather data obtained recorded for each city and it was downloaded from kaggle.

```python
df = pd.read_parquet('daily_weather.parquet', engine='fastparquet') 
df['date'] = pd.to_datetime(df['date']) # add as date
df['year'] = df['date'].dt.year  # add year columun
df.head()
```

<span style="font-size:20px; font-weight:bold">2. Data visualization</span></br> Next we will visiualize some of the data as follows. First, the temprature distribution of summer and winter seasons are plotted.

```python
summer_data = df[df['season'] == 'Summer']
winter_data = df[df['season'] == 'Winter']
plt.figure(figsize=(10, 6))
plt.hist(summer_data['avg_temp_c'].dropna(), bins=20, alpha=0.5, label='Summer', edgecolor='black')
plt.hist(winter_data['avg_temp_c'].dropna(), bins=20, alpha=0.5, label='Winter', edgecolor='black')
plt.title('Temperature Distribution by Season')
plt.xlabel('Temperature (°C or °F)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="font-size:15px; font-weight:bold">Analysis</span></br>The Summer Distribution is likely unimodal (has one peak) and the data is clustered at higher temperatures (e.g., 25–35°C).
Possible few cooler days are observed as per the data below 20°C
Represents the general warm climate during summer.
Where as the Winter Distribution has more clustered at lower temperatures (e.g., 0–10°C). The tail at the right ends show occasional mild or warm days pulling the tail rightward. This reflects cold conditions with some mixed in warmer days.
The histograms overlap significantly showing that there’s a range of temperatures that can occur in both seasons (e.g., spring or fall-like days).

```python
# Calculate yearly average temperature for each season
summer_trend = summer_data.groupby('year')['avg_temp_c'].mean().reset_index()
winter_trend = winter_data.groupby('year')['avg_temp_c'].mean().reset_index()
plt.figure(figsize=(12, 6))
# Plot Summer trend
plt.plot(
    summer_trend['year'],
    summer_trend['avg_temp_c'],
    marker='o',
    color='orange',
    linewidth=2,
    label='Summer'
)

# Plot Winter trend
plt.plot(
    winter_trend['year'],
    winter_trend['avg_temp_c'],
    marker='s',
    color='skyblue',
    linewidth=2,
    label='Winter'
)

plt.title('Average Summer and Winter Temperature Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.legend(title='Season')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="font-size:14px; font-weight:bold">Analysis</span></br> 
We can see the above plots that
1. Winter warming is more pronounced than summer warming which is commonly visible in climate research (especially at higher latitudes).
2. The decreasing gap between summer and winter temperatures points to less seasonal variability and possibly more mild winters in recent decades.
3. The general upward trend aligns with global climate change expectations. However, local factors (data collection methods and technologies, regional climate) could highly influence this trend as well

<span style="font-size:16px; font-weight:bold">3.1. Handling missing values</span></br> Droping NAN values have been shown during the historgram visualization part 2 above.
     We can also see from the summer and winner average temprature data trend that the most of the missing values exist before year 1880.

```python
mean_temp_summer = summer_data['avg_temp_c'].mean()
mean_temp_winter = winter_data['avg_temp_c'].mean()

summer_data_mean_replaced = summer_data.copy()
summer_data_mean_replaced = summer_data_mean_replaced.fillna(summer_data_mean_replaced.mean(numeric_only=True))

winter_data_mean_replaced = winter_data.copy()
winter_data_mean_replaced['avg_temp_c'] = winter_data_mean_replaced['avg_temp_c'].fillna(mean_temp_winter)
print("Summary Before Cleaning:")
print(summer_data.head())

print("\nSummary After Cleaning:")
print(summer_data_mean_replaced.head())
```

```python
plt.figure(figsize=(12, 6))

# Summer histogram
plt.hist(
    summer_data_mean_replaced['avg_temp_c'],
    bins=20,
    alpha=0.5,
    label='Summer (Mean Replaced)',
    color='orange',
    edgecolor='black'
)

# Winter histogram
plt.hist(
    winter_data_mean_replaced['avg_temp_c'],
    bins=20,
    alpha=0.5,
    label='Winter (Mean Replaced)',
    color='skyblue',
    edgecolor='black'
)

plt.title('Temperature Distribution by Season (Mean Replacement)')
plt.xlabel('Avg Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Using mean replacement, we can clearly see the frequent average tempratures for each group clearly. Why?
The mean replacement ensures no missing data. The data is now consistent in structure (no gaps). and c
However, it artificially inflates the frequency of the mean temperature.
The histogram’s shape is distorted because real variability (natural distribution) is smoothed over.
This biases interpreting the frequency of temperatures in statistical analysis.

<span style="font-size:16px; font-weight:bold">3.2 Outliner detection.</span></br>
We will detect the outliners per city and plot them 

```python
outliers_iqr_city_from_original = pd.DataFrame()

for city in df['city_name'].unique():
    city_data = df[df['city_name'] == city]
    for season in ['Summer', 'Winter']:
        city_season_data = city_data[city_data['season'] == season]
        if not city_season_data.empty:
            Q1 = city_season_data['avg_temp_c'].quantile(0.25)
            Q3 = city_season_data['avg_temp_c'].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = city_season_data[
                (city_season_data['avg_temp_c'] < lower_bound) |
                (city_season_data['avg_temp_c'] > upper_bound)
            ].copy()
            outliers['lower_bound'] = lower_bound
            outliers['upper_bound'] = upper_bound

            outliers_iqr_city_from_original = pd.concat(
                [outliers_iqr_city_from_original, outliers],
                ignore_index=True
            )
# Overall yearly average temperature
yearly_avg_temp_all = df.groupby(['year', 'season'], observed=False)['avg_temp_c'].mean().reset_index()

# Yearly average of city-specific outliers
yearly_outlier_avg_city_from_original = outliers_iqr_city_from_original.groupby(
    ['year', 'season'], observed=False)['avg_temp_c'].mean().reset_index()
print("Yearly average of temprature data outliners gouped by year and season")
print(yearly_outlier_avg_city_from_original.head(20))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

# Plot yearly average for all data
for season, color in zip(['Summer', 'Winter'], ['orange', 'skyblue']):
    season_all = yearly_avg_temp_all[yearly_avg_temp_all['season'] == season]
    plt.plot(
        season_all['year'],
        season_all['avg_temp_c'],
        linestyle='--',
        color=color,
        label=f'{season} Overall Avg'
    )

# Plot yearly average for outliers
for season, color in zip(['Summer', 'Winter'], ['red', 'blue']):
    season_outliers = yearly_outlier_avg_city_from_original[
        yearly_outlier_avg_city_from_original['season'] == season]
    plt.plot(
        season_outliers['year'],
        season_outliers['avg_temp_c'],
        marker='o',
        linewidth=2,
        color=color,
        label=f'{season} Outlier Avg'
    )

plt.title('Yearly Average Temperatures vs. Outlier Averages per City (Original Data, IQR Method)')
plt.xlabel('Year')
plt.ylabel('Avg Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="font-size:14px; font-weight:bold">Analysis</span></br> 
During winter, the outlier averages are below the actual (overall) average before 1940s. Most detected outliers in this period are cold extremes. This possibly reflects include a colder climate period, less urban heat island effect, and fewer warm extremes (heatwaves).
However, the outlier averages move closer to or align with the overall average. Detected outliers now include both cold snaps and heatwaves, but heatwaves might have become more frequent or severe. An overall warming trend is shown after 1940 which posibbility aligns with most reasearch results.
When It comes to the summer temprature, the outlines are close to the the real averages and show increasing trend with time, indincating that the cities are experiencing more warmer tempratures.

<span style="font-size:16px; font-weight:bold">3.3 Sampling and Dimension Reduction:</span></br>
We will do purpose driven sampling and dimension reduction here. Since we are doing more of summer and winter average temprature trend anlysis, the sampling done is 10% of the data for each citiy-year. As a result we should be able to see if the trend with remain intact.

```python
sampled_df = df.groupby(['city_name', 'year'], observed=False, group_keys=False).apply(lambda x: x.sample(frac=0.1, random_state=42),include_groups=True).reset_index(drop=True)

# Step 2: Keep essential columns only
columns_to_keep = ['city_name', 'year', 'season', 'avg_temp_c']
reduced_df = sampled_df[columns_to_keep]
print("Summary Before Sampling and Reduction:")
print(df.head())

print("\nSummary After Sampling and Reduction:")
print(reduced_df.head())
```

```python
# 3️⃣ Calculate yearly average trend of the sampled data
yearly_avg_sampled = reduced_df.groupby(['year', 'season'], observed=False)['avg_temp_c'].mean().reset_index()

# 4️⃣ Plotting
plt.figure(figsize=(14, 7))

# Plot the average trend by season
for season, color in zip(['Summer', 'Winter'], ['orange', 'skyblue']):
    season_avg = yearly_avg_sampled[yearly_avg_sampled['season'] == season]
    plt.plot(
        season_avg['year'],
        season_avg['avg_temp_c'],
        marker='o',
        linewidth=2,
        color=color,
        label=f'{season} Avg Temp (Sampled Data)'
    )

plt.title('Yearly Average Temperature Trend (Sampled Data, 10% Stratified by City and Year)')
plt.xlabel('Year')
plt.ylabel('Avg Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="font-size:14px; font-weight:bold">Analysis</span></br> 
We can see from the above graph that with proper sampling and unwanted columns reduction, we are able to plot the average temprature trend over the years. And the trends are similar to what we plotted earlier.

<span style="font-size:16px; font-weight:bold">3.4.Data Scaling and Discretization</span></br> 
Z-score Standardization and Discretize continuous data into meaningful categories. 

```python
non_scaled_yearly_avg = df.groupby('year', observed=False).agg({
    'avg_temp_c': 'mean',
    'precipitation_mm': 'mean'
}).reset_index()
yearly_avg = non_scaled_yearly_avg.copy()
scaler = StandardScaler()
yearly_avg[['temp_zscore', 'precip_zscore']] = scaler.fit_transform(
    yearly_avg[['avg_temp_c', 'precipitation_mm']]
)
print("Non scaled yearly average data")
print(non_scaled_yearly_avg.dropna().head(10))
print("Z-scaled yearly average data")
print(yearly_avg.dropna().head(10))
```

```python

plot_data = yearly_avg.dropna(subset=['temp_zscore', 'precip_zscore'])
plt.figure(figsize=(14, 7))

# Plot standardized temperature trend
plt.plot(
    plot_data['year'],
    plot_data['temp_zscore'],
    marker='o',
    linewidth=2,
    color='red',
    label='Temperature (Z-score)'
)

# Plot standardized precipitation trend
plt.plot(
    plot_data['year'],
    plot_data['precip_zscore'],
    marker='s',
    linewidth=2,
    color='blue',
    label='Precipitation (Z-score)'
)

plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Mean line

plt.title('Standardized Variability of Temperature and Precipitation Over Years')
plt.xlabel('Year')
plt.ylabel('Z-score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="font-size:14px; font-weight:bold">Analysis</span></br> 
The increase in temperature Z-score with time indicates that the cities in the dataset are warming over time.
This aligns with broader trends of global warming and urban heat island effects.
Comparing the temprature Z score, precipitation Z-score does not change that much after 1940. And it suggests that rainfall is not increasing proportionally to temperature. This could indicate a drier climate trend, especially if higher temperatures cause more evaporation.

```python
q1_temp = yearly_avg['avg_temp_c'].quantile(0.25)
q2_temp = yearly_avg['avg_temp_c'].quantile(0.50)
q3_temp = yearly_avg['avg_temp_c'].quantile(0.75)

# Define bins for temperature
bins_temp = [-np.inf, q1_temp, q2_temp, q3_temp, np.inf]
labels_temp = ['Q1 (Coolest)', 'Q2', 'Q3', 'Q4 (Warmest)']
yearly_avg['temp_category'] = pd.cut(yearly_avg['avg_temp_c'], bins=bins_temp, labels=labels_temp)

# Calculate quartiles for precipitation
q1_precip = yearly_avg['precipitation_mm'].quantile(0.25)
q2_precip = yearly_avg['precipitation_mm'].quantile(0.50)
q3_precip = yearly_avg['precipitation_mm'].quantile(0.75)

# Define bins for precipitation
bins_precip = [-np.inf, q1_precip, q2_precip, q3_precip, np.inf]
labels_precip = ['Q1 (Driest)', 'Q2', 'Q3', 'Q4 (Wettest)']
yearly_avg['precip_category'] = pd.cut(yearly_avg['precipitation_mm'], bins=bins_precip, labels=labels_precip)

print("Discretize continuous temprature and pericipitation data")
print(yearly_avg[['temp_category','precip_category']].dropna().head(15))
```

```python
heatmap_data = yearly_avg.dropna(subset=['temp_category', 'precip_category'])

# 2️⃣ Cross-tabulation to see the relation between temp and precip categories
cross_tab = pd.crosstab(heatmap_data['temp_category'], heatmap_data['precip_category'])

# 3️⃣ Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(
    cross_tab, 
    annot=True, 
    fmt='d', 
    cmap='coolwarm', 
    cbar_kws={'label': 'Count'}
)
plt.title('Heatmap of Temperature vs. Precipitation Categories (Quartiles)')
plt.xlabel('Precipitation Category')
plt.ylabel('Temperature Category')
plt.tight_layout()
plt.show()
```

<span style="font-size:14px; font-weight:bold">Analysis</span></br> 
Before 1940 tempratures were colder and below the logterm average. Same as percipitation but it is not dramatically low as the temprate. 
But in later times, when the temprature is increasing above longterm average, the rain faill also increases which in general shows that the warmesset years have more percipitation. This aligns with the Z-score plot we did earlier.

<span style="font-size:16px; font-weight:bold"> 4.3 Dispersion measures</span></br>

```python
# Temperature Dispersion
print("📊 Temperature Dispersion:")
temp_min = df['avg_temp_c'].min()
temp_max = df['avg_temp_c'].max()
temp_range = temp_max - temp_min
temp_q1 = df['avg_temp_c'].quantile(0.25)
temp_q2 = df['avg_temp_c'].quantile(0.50)
temp_q3 = df['avg_temp_c'].quantile(0.75)
temp_iqr = temp_q3 - temp_q1
temp_variance = df['avg_temp_c'].var()
temp_std = df['avg_temp_c'].std()

print(f"Range: {temp_range:.2f}")
print(f"Q1 (25th percentile): {temp_q1:.2f}")
print(f"Q2 (Median): {temp_q2:.2f}")
print(f"Q3 (75th percentile): {temp_q3:.2f}")
print(f"IQR: {temp_iqr:.2f}")
print(f"Variance: {temp_variance:.2f}")
print(f"Standard Deviation: {temp_std:.2f}")

# Precipitation Dispersion
print("\n🌧️ Precipitation Dispersion:")
precip_min = df['precipitation_mm'].min()
precip_max = df['precipitation_mm'].max()
precip_range = precip_max - precip_min
precip_q1 = df['precipitation_mm'].quantile(0.25)
precip_q2 = df['precipitation_mm'].quantile(0.50)
precip_q3 = df['precipitation_mm'].quantile(0.75)
precip_iqr = precip_q3 - precip_q1
precip_variance = df['precipitation_mm'].var()
precip_std = df['precipitation_mm'].std()

print(f"Range: {precip_range:.2f}")
print(f"Q1 (25th percentile): {precip_q1:.2f}")
print(f"Q2 (Median): {precip_q2:.2f}")
print(f"Q3 (75th percentile): {precip_q3:.2f}")
print(f"IQR: {precip_iqr:.2f}")
print(f"Variance: {precip_variance:.2f}")
print(f"Standard Deviation: {precip_std:.2f}")
```
